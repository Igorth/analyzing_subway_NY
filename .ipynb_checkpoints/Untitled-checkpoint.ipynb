```python
import numpy as np
import pandas
import matplotlib.pyplot as plt

df_rain = teste[teste.rain != 0]
df_norain = teste[teste.rain == 0]

df_rainent = df_rain['ENTRIESn_hourly']
df_norainent = df_norain['ENTRIESn_hourly']

plt.figure()
#df_norain.hist()
#df_rainent.hist()

   
teste['ENTRIESn_hourly'][teste.rain == 1].hist(bins = 25, range=(0,6000)) # raining, includes bins and range args
teste['ENTRIESn_hourly'][teste.rain == 0].hist(bins = 25,alpha = .4,range = (0,6000)) # not raining includes bins, range, and alpha transparency args
plt.xlabel("ENTRIESn_hourly")
plt.ylabel('Frequency')
plt.legend(('Rain','No Rain'))
plt.title('Ridership Distribution for Rain and No Rain Conditions')

plt.show()
```
